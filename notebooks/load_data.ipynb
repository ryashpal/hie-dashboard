# Ingest data to FHIR

## Patients

```python
import os
from pathlib import Path

import pandas as pd


save_dir = Path(os.environ['HIE_DASHBOARD_BASE'], 'data', 'v_1.0')
patients_file = Path(save_dir, 'patients.csv')

patients_df = pd.read_csv(patients_file)
patients_df
```

```python
patient_resources = []

# Iterate through each patient and create a FHIR Patient resource
for _, row in patients_df.iterrows():
    patient_resource = {
        "resourceType": "Patient",
        "id": row["patient_id"],
        "gender": row["gender"].lower(),
        "birthDate": row["dob"],
    }
    patient_resources.append(patient_resource)

print(len(patient_resources))
```

```python
import requests


# FHIR server URL
fhir_server_url = "http://localhost:80/fhir/Patient"

for patient_resource in patient_resources:
    response = requests.post(fhir_server_url, json=patient_resource, verify=False)
    if response.status_code != 201:
        print(f"Failed to create Patient resource: {response.status_code} - {response.text}")
```
